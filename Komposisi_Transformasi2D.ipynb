```python
import matplotlib.pyplot as plt
import numpy as np
import math

# --- Definisi titik segitiga ---
triangle = np.array([[0, 0, 1],
                     [2, 0, 1],
                     [1, 2, 1],
                     [0, 0, 1]]).T   # Transpose supaya setiap kolom = titik

# --- Fungsi untuk membuat matriks rotasi homogen ---
def rotation_matrix(theta_deg):
    theta = math.radians(theta_deg)
    return np.array([
        [math.cos(theta), -math.sin(theta), 0],
        [math.sin(theta),  math.cos(theta), 0],
        [0, 0, 1]
    ])

# --- Plot awal ---
plt.figure(figsize=(7,7))
plt.plot(triangle[0], triangle[1], 'b-', label='Segitiga Asli')

# --- Rotasi setiap 15°, mulai dari 30° hingga 360° ---
angle = 30
while angle <= 360:
    R = rotation_matrix(angle)
    rotated = R @ triangle
    plt.plot(rotated[0], rotated[1], '--', label=f'Rotasi {angle}°')
    angle += 15

# --- Tampilan plot ---
plt.title("Komposisi Rotasi Segitiga (30° hingga 360°, tiap +15°)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Fungsi untuk membuat titik-titik bintang 5 sudut
def create_star_points(radius_outer=5, radius_inner=2.5, n=5):
    points = []
    angle = math.pi / 2  # mulai dari atas
    step = math.pi / n  # sudut antar titik

    for i in range(2 * n):
        r = radius_outer if i % 2 == 0 else radius_inner
        x = r * math.cos(angle)
        y = r * math.sin(angle)
        points.append((x, y))
        angle += step
    points.append(points[0])  # tutup bentuk bintang
    return np.array(points)

# Fungsi untuk rotasi
def rotate(points, theta_deg):
    theta = math.radians(theta_deg)
    R = np.array([
        [math.cos(theta), -math.sin(theta)],
        [math.sin(theta),  math.cos(theta)]
    ])
    return np.dot(points, R.T)

# Titik awal bintang
star = create_star_points()

# Plot hasil rotasi dengan sudut istimewa (0°, 30°, 45°, 60°, 90°)
angles = [0, 30, 45, 60, 90]
colors = ['blue', 'red', 'green', 'orange', 'purple']
plt.figure(figsize=(7,7))
for i, angle in enumerate(angles):
    rotated_star = rotate(star, angle)
    plt.plot(rotated_star[:,0], rotated_star[:,1], color=colors[i], label=f'Rotasi {angle}°')

plt.title("Rotasi Bintang pada Sudut-Sudut Istimewa (0°, 30°, 45°, 60°, 90°)")
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()

```
