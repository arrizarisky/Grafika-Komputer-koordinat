```python
import matplotlib.pyplot as plt

# Persegi
x = [1, 3, 3, 1, 1]
y = [1, 1, 3, 3, 1]

# Skala (sx=2, sy=0.5)
sx, sy = 2, 0.5
x_s = [i * sx for i in x]
y_s = [j * sy for j in y]

plt.figure(figsize=(5,5))
plt.plot(x, y, 'b-', marker='o', label='Asli')
plt.plot(x_s, y_s, 'm--', marker='o', label='Skala (2,0.5)')
plt.legend(); plt.grid(True); plt.axis("equal")
plt.title("Skala Persegi")
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# Fungsi untuk menggambar rumah sederhana
def draw_house(ax, x_shift=0, y_shift=0, scale=1, rotation=0, color='blue', label='Asli'):
    # Titik rumah (dinding: persegi panjang, atap: segitiga, pintu: persegi panjang kecil)
    wall_x = [0, 4, 4, 0, 0]
    wall_y = [0, 0, 3, 3, 0]

    roof_x = [0, 2, 4, 0]
    roof_y = [3, 5, 3, 3]

    door_x = [1.5, 2.5, 2.5, 1.5, 1.5]
    door_y = [0, 0, 1.5, 1.5, 0]

 # Gabungkan bentuk
    shapes = [(wall_x, wall_y), (roof_x, roof_y), (door_x, door_y)]

    for x_points, y_points in shapes:
        new_x, new_y = [], []
        for (x, y) in zip(x_points, y_points):
            # Skala
            x, y = x * scale, y * scale
            # Rotasi
            rad = math.radians(rotation)
            xr = x * math.cos(rad) - y * math.sin(rad)
            yr = x * math.sin(rad) + y * math.cos(rad)
            # Translasi
            xr, yr = xr + x_shift, yr + y_shift
            new_x.append(xr)
            new_y.append(yr)
        ax.plot(new_x, new_y, marker='o',
color=color, label=label)
        label = None  # biar label cuma muncul sekali di legend


# Plot semua transformasi
fig, ax = plt.subplots(figsize=(6,6))

# Rumah asli
draw_house(ax, color='blue', label='Asli')

# Translasi (geser kanan & atas)
draw_house(ax, x_shift=6, y_shift=2, color='red', label='Translasi')

# Skala (diperbesar 1.5x)
draw_house(ax, scale=1.5, x_shift=12, color='green', label='Skala')

# Rotasi 45 derajat
draw_house(ax, rotation=45, x_shift=18, color='purple', label='Rotasi 45Â°')

ax.set_title("Transformasi Geometri 2D pada Gambar Rumah")
ax.legend()
ax.set_aspect('equal')
ax.grid(True)
plt.show()

```
