```python
import matplotlib.pyplot as plt

def dda(x1, y1, x2, y2):
  x, y = x1, y1
  dx = x2 - x1
  dy = y2 - y1
  steps = int(max(abs(dx), abs(dy)))
  x_inc = dx / steps
  y_inc = dy / steps

  xs, ys = [], []
  for i in range(steps + 1):
    xs.append(round(x))
    ys.append(round(y))
    x += x_inc
    y += y_inc

  return xs, ys
x, y = dda(3, 7, 11, 15)
plt.figure(figsize=(5,5))
plt.scatter(x, y, color='red')
plt.plot(x, y, color='red', linewidth=1)
plt.title("Algoritma DDA")
plt.xlim(0, 20); plt.ylim(0, 20); plt.grid(True)
plt.show()
```
